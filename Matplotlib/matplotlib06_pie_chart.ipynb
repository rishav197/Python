### Matplotlib Tutorial : Pie Charts

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### **Draw a pie chart to track down home expenses**

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.pie(exp_vals, labels=exp_labels)
```

### **Draw a pie chart as perfect circle**

```python
plt.pie(exp_vals, labels=exp_labels, shadow=True)
plt.axis('equal')
plt.show()
```

### **Show percentages for every pie. Specify radius to increase chart size**

```python
plt.axis('equal') # To increase the size of pie chart this statement should be above plt.pie()
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct='%.2f%%', radius=1.5)
plt.show()
```

### **Explode**

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct='%0.2f%%', explode=[0,0,0,0.1,0.2])
plt.show()
```

### **Counterclock and angle properties**

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=90)
plt.show()
```
