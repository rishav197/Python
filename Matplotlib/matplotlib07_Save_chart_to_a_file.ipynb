### **Saving chart to a file using savefig**

* 'piechart.png' → name and format of the saved file.
* bbox_inches='tight' → trims extra whitespace around the plot.
* pad_inches=2 → adds 2 inches of padding around the figure.
* transparent=True → makes the background transparent in the saved image

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig('piechart.png', bbox_inches='tight', pad_inches=2, transparent=True)
```
