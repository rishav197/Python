## **Pandas Merge Tutorial**

### **Basic Merge Using a Dataframe column**

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'city': ["new york","chicago","orlando"],
    'temperature' : [21, 14, 35]
})

df1
```

```python
df2 = pd.DataFrame({
    'city': ["chicago","new york","orlando"],
    'humidity': [65, 68, 75]
})

df2
```

```python
df3 = pd.merge(df1, df2, on='city')
df3
```

### **Types of Database Joins**

<img src="https://github.com/rishav197/Python/blob/main/Pandas/pandas09_pandas_merge/db_joins.jpg?raw=true" width="400">

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})

df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})

df2
```

**INNER JOIN**

```python
df3 = pd.merge(df1, df2, on='city', how='inner')
df3
```

**OUTER JOIN**

```python
df3 = pd.merge(df1, df2, on='city', how='outer')
df3
```

**LEFT JOIN**

```python
df3 = pd.merge(df1, df2, on='city', how='left')
df3
```

**RIGHT JOIN**

```python
df3 = pd.merge(df1, df2, on='city', how='right')
df3
```

### **Indicator flag**

```python
df3 = pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

### **Suffixes**

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})

df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})

df2
```

```python
df3 = pd.merge(df1,df2,on="city",how="outer", suffixes=('_left','_right'))
df3
```

### **Join**

1. .join() is a Pandas DataFrame method used to combine two DataFrames side by side (column-wise), usually aligning them by the index.

2. It’s similar to a SQL join but by default it works on the index, unless you specify otherwise.

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "LA"],
    "temperature": [21,14,35, 40],
})

df1.set_index('city',inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})

df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2, lsuffix='_l', rsuffix='_r')
```
