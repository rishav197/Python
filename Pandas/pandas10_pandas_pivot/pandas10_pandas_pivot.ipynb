### **Pivot basics**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/pandas10_pandas_pivot/weather.csv')
df
```

```python
df.pivot(index='city', columns='date')
```

### **Pivot Table**

Pivot table is used to summarize and aggregate data inside dataframe

```python
df = pd.read_csv('/content/drive/MyDrive/pandas10_pandas_pivot/weather2.csv')
df
```

```python
df.pivot_table(index='city', columns='date')
```

#### **Margins**

```python
df.pivot_table(index='city', columns='date', margins=True, aggfunc='sum')
```

```python
df.pivot_table(index='city', columns='date', margins=True, aggfunc='mean')
```

### **Grouper**
