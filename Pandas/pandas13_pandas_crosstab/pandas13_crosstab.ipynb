### **Crosstab Tutorial**

#### Before we dive into all things crosstabs, it's worth getting our definitions right. A crosstab (short for cross-tabulation) is a table used in statistics to display the relationship between two or more categorical variables. It allows you to examine how different categories or groups of variables intersect and how they are distributed within a dataset. In short, a crosstab summarizes categorical data, providing a quick and easy way to see patterns or associations between variables.

```python
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/pandas13_pandas_crosstab/survey.xls')
df
```

```python
pd.crosstab(df['Nationality'], df['Handedness'])
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

```python
pd.crosstab(df['Sex'], df['Handedness'])
```

### **Margins**

```python
pd.crosstab(df['Sex'], df['Handedness'], margins=True)
```

### **Multi Index Column and Rows**

```python
pd.crosstab(df['Sex'], [df['Handedness'], df['Nationality']], margins=True)
```

```python
pd.crosstab([df['Nationality'], df['Sex']], df['Handedness'], margins=True)
```

### **Normalize**

```python
pd.crosstab(df['Sex'], df['Handedness'], normalize='index')
```

### **Aggfunc and Values**

```python
import numpy as np
```

```python
pd.crosstab(df['Sex'], df['Handedness'], values=df['Age'], aggfunc=np.average)
```
