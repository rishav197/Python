### **What is DatetimeIndex? Benefits of it**

##### (1) Partial Date Index: Select Specific Months Data

```python
df.loc['2017-07-07']
```

```python
print(len(df.loc['2017-01']))
df.loc['2017-01']
```

```python
df.loc['2017-06'].head()
```

##### Average price of aapl's stock in June, 2017

```python
df.loc['2017-06']['Close'].mean()
```

### **Resampling**

##### Resample your Close prices by month-end,<br> Take the mean within each month,<br> Plot a bar chart of monthly average closing prices.

```python
df['Close'].resample('ME').mean().head()
```

```python
%matplotlib inline
df['Close'].plot()
```

```python
df['Close'].resample('ME').mean().plot()
```

```python
df['Close'].resample('ME').mean().plot(kind='bar')
```
