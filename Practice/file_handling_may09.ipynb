```python
# Reading one line at a time
file = open('/content/drive/MyDrive/sample_text/sample_test.txt', 'r')
line = file.readline()
print(line)
file.close()
```

```python
# Reading one line at a time
file = open('/content/drive/MyDrive/sample_text/sample_test.txt', 'r')

for line in file:
  print(line.strip())

file.close()
```

```python
file = open('/content/drive/MyDrive/sample_text/sample_test.txt', 'r')

lines = file.readlines()

for line in lines:
  print(line.strip())

file.close()
```

3. Close a File

Instead of manually opening and closing a file, you can use the with statement, which
automatically handles closing the file when the block of code is done.

```python
# Reading with with statement
with open('/content/drive/MyDrive/sample_text/sample_test.txt', 'r') as file:
  content = file.read()
  print(content)
```

```python
# Example: Using exception handling to close a file
try:
  with open('/content/drive/MyDrive/sample_text/sample_test.txt', 'r') as file:
    content = file.read()
    print(content)
finally:
  file.close()
```
