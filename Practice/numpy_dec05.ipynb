We will add two list of numbers using python list and numpy array and compare them

```python
size = 100

lst1 = range(size)
lst2 = range(size)

# python list
start = time.time()
result = [(x+y) for x,y in zip(lst1,lst2)]
print('Python list took :', (time.time()-start)*1000)
```

```python
a1 = np.arange(size)
a2 = np.arange(size)

#numpy array
start = time.time()
result = a1 + a2
print('numpy array took :', (time.time()-start)*1000)
```

### **Reshape**

```python
a = np.array([[1,2], [3,4], [5,6]])
print(a)
print(a.shape)
print('\n')

a = a.reshape(2, 3)
print(a)
print(a.shape)
print('\n')

a = a.reshape(6, 1)
print(a)
print(a.shape)
print('\n')

a = a.ravel()
print(a)
print(a.shape)
print('\n')
```

Iterate through arrays

```python
a = np.array([[6, 7, 8], [1, 2, 3], [9, 1, 5]])
a
```

```python
# Iterating the complete matrix/numpy array
for row in a:
  # print(row)
  lst = row
  for i in lst:
    print(i, end=' ')
  print()
```

```python
for num in a.flat:
  print(num)
```

```python
for num in a.ravel():
  print(num)
```

### **Iterate Numpy array using nditer**

```python
a = np.arange(12).reshape(3, 4)
a
```

```python
for row in a:
  for cell in row:
    print(cell)
```

```python
for cell in a.flatten():
  print(cell)
```

### **Numpy function nditer() function**

```python
for x in np.nditer(a, order='C'):
  print(x)
```

```python
for x in np.nditer(a, order='F'):
  print(x)
```

```python
for x in np.nditer(a, order='C', flags=['external_loop']):
  print(x)
```

### Op_flags on for loop

```python
for x in np.nditer(a, op_flags=['readwrite']):
  x[...] = x*x

a
```
