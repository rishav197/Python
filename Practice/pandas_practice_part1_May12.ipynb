```python
def convert_people_cell(cell):
  if cell=='n.a.':
    return 'Sam Walton'

  return cell

def convert_price_cell(cell):
  if cell=='n.a.':
    return 50

  return cell
```

Write two dataframes to two separate sheets in excel

```python
df_stocks = pd.DataFrame({
    'tickers':['GOOGL', 'WMT', 'MSFT'],
    'price':[845, 65, 64],
    'pe':[30.37, 14.26, 30.97],
    'eps':[27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
  df_stocks.to_excel(writer, sheet_name='stocks', index=False)
  df_weather.to_excel(writer, sheet_name='weather', index=False)
```

```python
with pd.ExcelWriter('weather_stocks.xlsx') as writer:
  df_weather.to_excel(writer, sheet_name='weather', index=False)
  df_stocks.to_excel(writer, sheet_name='stocks', index=False)
```

Handling Missing Data - Fillna, interpolate, dropna

```python
# parse_dates=['day'] tells Pandas to automatically convert the day column into a datetime datatype while reading the CSV file
df = pd.read_csv('/content/drive/MyDrive/pandas05_handling_missing_data/weather_data.csv', parse_dates=['day'])
df
```

```python
df.info()
```

fillna

Fill all NaN/Null values with one specific value


```python
df.fillna(10)
```

Fill na using column names and dictionary

```python
df.fillna({
    'temperature':0,
    'windspeed':0,
    'event': 'No Event'
})
```

Use method to determine how to fill na values

```python
# forward fill
df.fillna(method='ffill')
```

```python
# backward fill
df.fillna(method='bfill')
```

```python
df.fillna(method='ffill', limit=1)
```

```python
df.fillna(method='bfill', axis='columns') # axis is either 'index' or 'columns'
```

Interpolate

Estimate missing values using nearby known values.

```python
df = df.set_index('day')
df
```

```python
df.interpolate(method='time')
```

dropna

```python
df.dropna()
```

```python
# Removes rows where all values are missing (NaN).
df.dropna(how='all')
```

```python
# 'thresh' is Minimum number of non-NaN values required to keep a row.
df.dropna(thresh=1)
```
