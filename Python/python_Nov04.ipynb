### Tuples

They are like lists (sequence of objects) but they are immutable i.e. once they have been defined we cannot change them. Parenthesis in tuples are optional.

```python
marks = (95, 98, 97, 97, 97, 97)
#marks[0] = 98 # tuple is immutable
```

```python
print(marks.count(97))
print(marks.index(95))
print(marks.index(98))
print(marks.index(97))
```

```python
print(marks[:4])
print(marks[2])
```

### Dictionary

Dictionary is an unordered collection of items. Dictionary stores a (key, value) pair.

```python
marks = {
    "math" : 99,
    "chemistry" : 98,
    "physics" : 97
}
```

```python
print(marks)
print(marks["chemistry"])
print(marks["math"])
```

Iterating through a dictionary

```python
# Iterate over keys
for key in marks:
  # print(key, marks[key])
  # print({key})
  print(f"{key}: {marks[key]}")

# OR

# Iterate over keys
for key in marks.keys():
  print(key)
```

```python
# Iterate over values
for value in marks.values():
  print(value)
```

```python
# Iterate over key-value pairs
for key, val in marks.items():
  print(f"{key}: {val}")
```

```python
# Nested Dictionary

d = {1: 'Geeks', 2: 'For',
        3: {'A': 'Welcome', 'B': 'To', 'C': 'Geeks'}}

print(d)
```
