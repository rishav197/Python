We will add two list of numbers using python list and numpy array and compare them

```python
import time
```

```python
size = 1000

l1 = range(size)
l2 = range(size)

# python list
start = time.time()
result = [(x+y) for x,y in zip(l1, l2)]

# print(result)
print(f"Time taken by Python List : {(time.time()-start)*1000}")
```

```python
size = 1000

a1 = np.arange(size)
a2 = np.arange(size)

#numpy array
start = time.time()
result = a1 + a2

# print(result)
print(f"Time taken by Numpy Array : {(time.time()-start)*1000}")
```

Reshape

```python
a = np.array([[1,2], [3,4], [5,6]])
print(a)
print(a.shape)
print('\n')

a = a.reshape(2,3)
print(a)
print(a.shape)
print('\n')

a = a.reshape(6, 1)
print(a)
print(a.shape, a.ndim)
print('\n')

a = a.reshape(1, 6)
print(a)
print(a.shape, a.ndim)
print('\n')

a = a.reshape(6,)
print(a)
print(a.shape, a.ndim)
print('\n')
```

```python
a = np.array([[9,20], [13,14], [-5,61]])
print(a)
print(a.shape, a.ndim)
print('\n')

a = a.ravel() #flatten any array into 1D array
print(a)
print(a.shape, a.ndim)
```

Iterate through arrays

```python
a = np.array([[6, 7, 8], [1, 2, 3], [9, 3, 2]])
a
```

```python
for row in a:
  print(row)
```

```python
for row in a:
  for num in row:
    print(num, end=' ')
```

```python
for num in a.flat:
  print(num, end=' ')
```

```python
for num in a.ravel():
  print(num, end=' ')
```
